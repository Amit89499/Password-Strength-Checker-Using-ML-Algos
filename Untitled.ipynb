# Checking Password Strenth using ML Algorithms
The Dataset is taken from UCI Machine Learning Repository

```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv('data.csv',error_bad_lines=False) # offending lines are skipped
```

```python
data.head()
```

```python
data.shape
```

```python
data.isnull().sum()
```

```python
data.dropna(inplace=True)  # removing null values
```

```python
password_list=np.array(data)
password_list
```

```python
import random
random.shuffle(password_list)
```

```python
x=[labels[0] for labels in password_list]
y=[labels[1] for labels in password_list]
```

```python
import seaborn as sns
```

```python
#sns.set_style('whitegrid')
sns.countplot(x='strength',data=data,palette='RdBu_r')
```

```python
x
```

```python
y
```

```python
def word_divide_char(inputs):
    characters=[]
    for i in inputs:
        characters.append(i)
    return characters
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer=TfidfVectorizer(tokenizer=word_divide_char)
x=vectorizer.fit_transform(x)
```

```python
x.shape
```

```python
vectorizer.vocabulary_
```

```python
feature_names=vectorizer.get_feature_names()
first_document_vector=x[0]
df=pd.DataFrame(first_document_vector.T.todense(),index=feature_names,columns=["tfidf"])
df.sort_values(by=['tfidf'],ascending=False)
```

```python
# train test splitting
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=42)
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
classifier=LogisticRegression(penalty='l2',multi_class='ovr')
classifier.fit(X_train,y_train)
```

```python
print("Test accuracy {}".format(classifier.score(X_test,y_test)))
```

# xgboost 

```python
import xgboost as xgb
xgbClassifier=xgb.XGBClassifier()
xgbClassifier.fit(X_train,y_train)
```

```python
print(xgbClassifier.score(X_test,y_test))
```
